案例分析：

Not general: small input instance
eg. 在此教室中，有一对学生有相同生日么？

General: arbitrarily large inputs
eg.给定任意n个学生的集合，有一对学生有相同生日么？

算法：
程序映射每个输入到单个输出

如果算法对于每个问题输入都返回一个正确输出，则算法解决了问题


为保证算法的正确性与效率，以案例来分析;
正确性：
对于 small input instance 例子分析
对于 arbitrarily large inputs 递归or循环
必须要推断
eg.生日匹配算法的正确性证明可以分为以下步骤：

    基本情况（Base Case）： 当记录中没有学生时（即k=0），我们可以确定前k个学生不包含匹配。这是显而易见的，因为没有学生可以进行匹配。所以基本情况成立。

    归纳假设（Inductive Hypothesis）： 假设对于某个整数k=k'，如果前k'个学生包含匹配，则在采访第k'+1个学生之前，算法会返回匹配结果。或者如果前k'个学生不包含匹配，则前k'+1个学生中会包含匹配。

    归纳步骤（Inductive Step）： 考虑k=k'+1的情况。如果前k'个学生中包含匹配，根据归纳假设，算法在采访第k'+1个学生之前会返回匹配结果，因此我们已经找到了匹配。如果前k'个学生不包含匹配，那么第k'+1个学生的生日就是新的信息。如果第k'+1个学生的生日与前k'个学生的任何一个生日匹配，则算法会直接检测到匹配，并返回结果。如果第k'+1个学生的生日不与任何一个前k'个学生的生日匹配，则我们无法在采访第k'+1个学生之前得知匹配结果。

通过归纳证明，我们证明了在生日匹配算法中，如果存在匹配，算法会在采访第k个学生之前返回匹配结果。，归纳假设是为了推导出当k=k'时的情况成立，其中k'是任意的整数。通过假设这一情况成立，我们可以在推导k=k'+1时，根据这个假设来考虑两种可能性：一种是在前k'个学生中已经找到了匹配，另一种是前k'个学生中没有匹配，但在第k'+1个学生中找到了匹配。

效率：


    算法产生正确输出的速度评估：
        算法产生正确输出的速度：
        我们可以衡量算法的时间，但是我们希望性能独立于机器。
        我们想要知道算法需要多少个固定时间的操作来返回正确的输出。
        通常期望算法的运行时间随着输入大小的增加而增加，输入大小通常用'n'来表示，但也可能用其他符号。
        如果算法在与输入大小相关的多项式时间内返回结果，则认为它是高效的。
        有时候，对于某个问题，不存在有效的算法（见L20）。


    渐进符号
        忽略常量因子和低阶项：
        渐进分析关注的是当输入趋于无穷大时算法的行为。它忽略了常数因子和低阶项，因为在大输入规模下，这些因子相对于整体运行时间来说通常不那么重要。
        渐进符号的目的是在输入规模趋于无穷大时描述算法的行为。
        上界（O）表示算法的最坏情况运行时间的上限。
        下界（Ω）表示算法的最佳情况运行时间的下限。
        紧密界限（Θ）表示算法的运行时间在最佳和最坏情况下的紧密界限。
        时间估计通常基于假设每个操作需要一个时钟周期，这样我们可以将算法的运行时间与处理器的时钟周期相关联，以便更好地理解算法的相对速度。

    时钟周期的理解：
        假设我们有一个1GHz的单核处理器，这意味着处理器每秒钟可以执行10^9个时钟周期（即1 GHz = 10^9 次/秒）。
        当我们说算法的运行时间是一个操作一个时钟周期时，我们假设每个基本操作（比如加法、乘法、比较等）只需一个时钟周期来完成。
        通过将算法操作的数量除以处理器的时钟周期数量，我们可以得到一个大致的时间估计，以处理器时钟周期为单位。这样做是为了理解算法的相对速度，而不是依赖于特定机器的性能。

计算模型（Word-RAM）


机器字（Machine word）：Word-RAM模型中，机器字是一个固定大小的位块，通常用来表示一个整数或其他数据类型。这个位块的大小由参数w指定，称为“字长”（word size）。每个机器字有w位，例如，32位或64位。

内存（Memory）：内存是一个可寻址的机器字序列，用来存储数据和程序指令。每个机器字都有一个唯一的地址，可以通过这个地址来读取或写入内存中的数据。

处理器（Processor）：Word-RAM处理器支持许多常数时间（O(1)）内完成的操作，这意味着无论输入规模大小如何，操作的执行时间都是固定的。支持的操作包括整数算术运算（+、-、*、//、%）、逻辑运算（&&、||、!、==、<、>、<=、>=）、位运算（&、|、<<、>>等），以及读取和写入内存操作。

内存地址：内存地址必须能够访问内存中的每个位置。为了实现这一点，需要确保机器字的位数至少能够表示内存地址的位数，即log2 n，其中n是内存地址的最大数量。例如，如果最大内存地址需要32位来表示，那么机器字的大小至少应为32位，以便能够访问所有内存地址。

Python的计算模型：尽管Python是一种更复杂的计算模型，但它仍然可以在Word-RAM上实现。这意味着Python中的基本操作，如整数算术、逻辑运算等，可以被认为是在Word-RAM模型中以常数时间执行的操作。




数据结构：



数据结构（Data structure）：数据结构是一种存储非常量数据的方式，它支持一组操作。这些操作构成了数据结构的接口。

接口（Interface）：接口是一组定义了数据结构所支持操作的规范。在这里，文中提到了两种常见的接口：序列（Sequence）和集合（Set）。
    序列（Sequence）：序列中的数据项有一个外部顺序，可以根据它们的位置进行访问，例如访问第一个、最后一个或第n个数据项。
    集合（Set）：集合中的数据项具有内部顺序，查询操作基于数据项的键（key）而不是它们的位置。

数据结构的实现方式：不同的数据结构可以实现相同的接口，但它们的性能可能不同。例如，静态数组（Static Array）和动态数组（Dynamic Array）都可以实现序列接口，但它们的性能可能不同。

静态数组（Static Array）的示例：
    StaticArray(n)：分配大小为n的静态数组，并在Θ(n)的时间内初始化为0。
    StaticArray.get_at(i)：在Θ(1)的时间内返回数组索引i处存储的数据项。
    StaticArray.set_at(i, x)：在Θ(1)的时间内将数据项x写入数组索引i处。

存储的数据项可以是更大对象的地址：静态数组中存储的数据项可以是指向更大对象（如其他数据结构）的地址，而不仅仅是简单的数据项。

Python中的数据结构：文中提到了Python中的元组（tuple）和列表（list）。元组类似于静态数组，而列表则类似于动态数组，其性能取决于实现方式。





```python

class StaticArray:
    def __init__(self, size):
        self.size = size
        self.array = [None] * size

    def set_at(self, index, value):
        if 0 <= index < self.size:
            self.array[index] = value
        else:
            raise IndexError("Index out of range")

    def get_at(self, index):
        if 0 <= index < self.size:
            return self.array[index]
        else:
            raise IndexError("Index out of range")

    def __len__(self):
        return self.size

def birthday_match(students):
    '''
    Find a pair of students with the same birthday
    Input: tuple of student (name, bday) tuples
    Output: tuple of student names or None
    '''
    n = len(students) # O(1)
    record = StaticArray(n) # O(n)
    for k in range(n): # n
        (name1, bday1) = students[k] # O(1)
    # Return pair if bday1 in record
        for i in range(k): # k
            (name2, bday2) = record.get_at(i) # O(1)
            if bday1 == bday2: # O(1)
                return (name1, name2) # O(1)
        record.set_at(k, (name1, bday1)) # O(1)
    return None # O(1)
```
