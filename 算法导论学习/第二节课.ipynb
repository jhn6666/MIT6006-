第二节课：数据结构

计算机科学中，数据结构是用于储存和组织数据的方式，同时提供了对于数据执行操作的算法。
数据结构的设计是为了使得特定的操作能够以高效的方式执行，通常涉及到选择适当的数据组织方式以及涉及相应的算法

数据结构接口（称为API，即应用程序接口，or ADT，即抽象数据类型）定义了一组操作，这些操作描述了数据结构支持的功能。这些操作包括添加、删除、查找、访问等。接口提供了对数据结构的逻辑视图，即用户只需知道数据结构支持哪些操作，而不需要了解具体的实现细节。

两种主要的数据结构接口：

Sequence（序列）：序列是一种有序的数据集合，其中每个元素都有一个唯一的索引。序列接口支持以下操作：
    
访问（Access）：通过索引访问序列中的元素。通常可以通过索引来获取特定位置的元素值。例如，可以使用sequence[i]来访问序列中索引为i的元素。

插入（Insertion）：在序列中的指定位置插入新的元素。插入操作允许在序列的任意位置添加新的元素。通常，插入操作可能会导致其他元素的移动以腾出空间来插入新元素。

删除（Deletion）：删除序列中的指定位置的元素。删除操作允许从序列中删除特定位置的元素，并且通常会导致序列中其他元素的移动以填补删除的位置。

长度（Length）：获取序列中存储的元素数量。长度操作返回序列中元素的总数。

遍历（Traversal）：遍历序列中的所有元素。遍历操作允许按顺序访问序列中的每个元素，以便对每个元素执行特定的操作。


序列的常见实现包括数组、链表、栈、队列等。

Set（集合）：集合是一种无序的数据集合，其中不允许重复元素。集合接口支持以下操作：
    添加（Addition）：向集合中添加新元素。
    删除（Removal）：从集合中删除指定元素。
    包含（Containment）：检查集合中是否存在指定元素。
    长度（Length）：获取集合的元素数量。
    遍历（Traversal）：遍历集合中的所有元素。

集合的常见实现包括哈希集合、平衡树集合等。

1.Sequence接口：
用于维护一系列项目，其中项目的顺序是外在的（extrinsic），即项目按照特定的顺序排列。在Sequence接口中，通常假设序列中的每个项目都有一个唯一的索引，这个索引通常从0开始（即零索引）。序列的长度通常用n来表示，表示序列中存储的项目数量。

常见的实现Sequence接口的数据结构：

数组（Array）：使用连续的内存块来存储元素，可以通过索引进行快速访问。数组支持常数时间复杂度的访问操作，但在插入和删除操作时可能需要移动大量元素。

链表（Linked List）：由节点组成，每个节点包含一个元素和指向下一个节点的引用。链表支持在任意位置插入和删除元素，但访问操作的时间复杂度通常是线性的。

栈（Stack）：遵循后进先出（LIFO）原则的数据结构，只允许在栈顶进行插入和删除操作。

队列（Queue）：遵循先进先出（FIFO）原则的数据结构，只允许在队列的末尾进行插入操作，在队列的前端进行删除操作。


2.Set 接口：
Set接口允许我们以更高效的方式管理具有唯一键的项目集合。通常情况下，键可以是项目本身，但在某些情况下，可能需要使用额外的信息作为键来确保项目的唯一性。

常见的Set接口的实现包括：

哈希集合（Hash Set）：使用哈希表来存储项目，保证了快速的添加、删除和查找操作。

平衡树集合（Balanced Tree Set）：使用平衡树（如红黑树）来存储项目，保证了有序性的同时具有较快的添加、删除和查找操作。

位向量集合（Bit Vector Set）：使用位向量来表示集合中的元素，适用于有限范围内的整数键。

布隆过滤器（Bloom Filter）：用于快速检查一个元素是否可能存在于集合中，但存在一定的误判率。


1）Array Sequence指的是使用数组来实现Sequence接口。数组在静态操作方面表现出色，例如通过索引直接访问元素（get at(i)）和修改元素（set at(i, x)）的时间复杂度都是常数时间（Θ(1)）。这是因为数组在内存中是连续存储的，可以通过索引计算出元素在内存中的地址，因此访问和修改元素的时间是固定的。

然而，数组在动态操作方面表现不佳。当需要在数组中插入或删除元素时，可能需要执行以下操作：

    重新分配数组大小（Reallocating the array）：如果数组已满，并且需要插入新元素，则需要重新分配更大的内存空间来容纳新的元素。这通常涉及到申请一块更大的内存空间，并将原数组中的元素复制到新的内存空间中。

    移动修改位置后的所有元素（Shifting all items after the modified item）：如果在数组中间插入或删除元素，则需要将插入位置后的所有元素向后移动或将删除位置后的所有元素向前移动，以保持元素的连续性和顺序性。

这些操作会导致动态操作的时间复杂度增加，通常是线性的或更高的时间复杂度。这使得在大量插入或删除操作时，数组序列的性能可能不佳。

尽管数组在静态操作方面性能良好，但在需要频繁执行动态操作的情况下，可能需要考虑其他数据结构，如链表或动态数组，来更好地支持动态操作。链表具有在任意位置快速插入和删除元素的优势，而动态数组则通过动态调整数组大小来减少重新分配和移动元素的开销。



2）Linked List Sequence
链表是一种指针数据结构，其中每个项目存储在一个节点中，并且每个节点包含一个指向序列中下一个节点的指针。每个节点通常有两个字段：一个用于存储项目的字段（node.item），另一个用于指向下一个节点的指针（node.next）。

链表的优势在于可以通过简单地重新连接指针来操作节点，而不需要像数组那样重新分配内存或移动大量元素。通过维护指向序列中第一个节点的指针（称为头部），我们可以在链表的开头执行插入和删除操作，并且这些操作的时间复杂度为常数时间（Θ(1)）。

具体来说，链表的特点如下：

    在开头插入和删除（Insertion and Deletion from Front）：由于我们可以直接访问头部节点的指针，因此在链表的开头执行插入和删除操作的时间复杂度是常数时间（Θ(1)）。这使得链表非常适合作为栈或队列的实现。

    在末尾插入和删除（Insertion and Deletion from Back）：尽管在末尾插入和删除操作的时间复杂度也可以是常数时间，但需要维护对链表末尾节点的指针，这在某些情况下可能需要额外的操作。在某些链表实现中，尾部指针可以被用来实现高效的末尾操作。

    获取和设置（Get and Set）：由于链表是一种单向结构，并且不能直接通过索引访问元素，因此获取和设置操作的时间复杂度通常是线性的（O(n)），因为需要从头节点开始遍历链表直到达到目标位置。

虽然链表在动态操作方面表现出色，但在获取和设置操作方面性能较差。为了兼顾两者，我们可以考虑使用更高级别的数据结构，如跳表（Skip List）或双向链表（Doubly Linked List），它们可以在某些情况下提供更好的性能。

跳表：

跳表是一种基于链表的数据结构，它通过添加多层次的索引来加快搜索的速度。
在跳表中，每个节点除了指向下一个节点的指针之外，还有指向更远节点的指针，称为“跳跃指针”（skip pointers）。
跳表的每一层都是一个有序链表，最底层包含所有的节点，而上面的层次则包含底层的一部分节点。
跳表允许通过跳跃指针快速地定位节点，从而加速搜索、插入和删除操作的性能。
跳表的平均时间复杂度为O(log n)，其中n是元素的数量，这使得它在某些情况下比传统链表更高效。

双向链表：

双向链表是一种链表，每个节点都有两个指针，一个指向前一个节点，另一个指向后一个节点。
双向链表允许从头部或尾部以常数时间（Θ(1)）进行插入和删除操作，因为它们可以直接访问前一个或后一个节点。
双向链表还可以支持双向遍历，即可以从头部向后遍历或从尾部向前遍历。
双向链表在某些情况下比单向链表更方便，尤其是需要频繁在两个方向上移动的情况下，例如实现LRU（Least Recently Used）缓存算法。



3)动态数组序列（Dynamic Array Sequence）是一种使数组在动态操作方面更加高效的数据结构。它的设计思想是通过在数组的末尾预先分配一定量的额外空间，使得在进行动态操作时不需要每次都进行数组的重新分配。

具体来说，动态数组序列的特点如下：

    填充比率（Fill Ratio）：动态数组序列会维护一个填充比率r，它表示数组中实际存储的项目与数组总空间之间的比率。填充比率的范围通常是0 ≤ r ≤ 1。

    动态空间分配（Dynamic Space Allocation）：当数组的填充比率达到1时，即数组已经满载，动态数组序列会分配额外的空间来容纳新的元素。通常情况下，会分配Θ(n)个额外的空间，其中n是当前数组的大小。这个额外空间的分配可以保证在一定程度上减少重新分配的次数，从而提高性能。

    平均时间复杂度为Θ(1)：尽管单次重新分配操作可能需要Θ(n)的时间，但是通过预先分配额外空间，使得任意长度为Θ(n)的操作序列的平均时间复杂度仍然是Θ(1)。这是因为重新分配操作的次数相对较少，而在每次重新分配之后，可以执行多个动态操作而不需要进行额外的内存重新分配。

    高效的尾部操作：由于额外空间是在数组的末尾进行分配的，因此在末尾执行插入操作的性能非常高效，平均时间复杂度为常数时间（Θ(1)）。

Python中的“list”就是一种动态数组的实现，它通过类似的策略来提高动态操作的性能。这使得Python的列表在实践中非常高效，能够很好地应对各种动态操作。



4）摊销分析（Amortized Analysis）是一种数据结构分析技术，用于在多次操作中分配成本，以便更好地理解数据结构的整体性能。

    摊销成本（Amortized Cost）：在摊销分析中，我们关注操作的摊销成本T(n)，这表示进行k次操作的总成本最多不超过k乘以T(n)。这里的T(n)通常指的是某种平均意义上的操作成本，可以是均摊下来的实际操作时间或者其他资源消耗。

    摊销时间（Amortized Time）：如果某个操作具有Θ(1)的摊销时间，意味着在进行k次操作时，总成本不会超过k个常数时间的操作成本。换句话说，每次操作的摊销成本是常数级别的，即使有些操作可能会花费更多时间，但是总体上在多次操作中的平均成本是可控的。

    动态数组插入的摊销时间：例如，对于动态数组来说，在进行一次插入操作时，可能需要进行重新分配等操作，导致时间复杂度为Θ(n)。但是通过摊销分析，我们可以证明，在一系列插入操作中，每次插入的摊销时间仍然是Θ(1)的，这意味着动态数组插入操作的平均成本是常数级别的。

    更多摊销分析技术：摊销分析并不局限于动态数组，它是一种通用的分析技术，可以应用于各种数据结构和算法中。在《Introduction to Algorithms》中，以及其他算法课程中，通常会深入讨论不同数据结构和算法的摊销分析方法，以便更好地理解它们的性能特征和使用场景。

总的来说，摊销分析通过将成本分配到多个操作中，帮助我们更好地评估数据结构和算法的整体性能，特别是对于那些在单次操作可能具有较高成本但在平均情况下表现良好的情况下，摊销分析尤为有用。



5）动态数组的删除操作通常在末尾进行，时间复杂度为Θ(1)，这在一般情况下非常高效。然而，这种删除方式可能会造成空间的浪费，因为并未减少数组的大小，导致实际存储的元素数量与数组的总空间之间存在差距。为了解决这个问题，需要对动态数组的大小进行调整，以便在保持总体大小为Θ(n)的情况下，尽可能减少额外的空间占用。

以下是关于动态数组删除操作的一些策略和想法：

    基本思路：当动态数组中的元素数量很小时，可以考虑将数组的大小调整为实际元素数量，即 r = 1。这样可以避免不必要的空间浪费。

    动态调整大小：为了在维持总体大小为Θ(n)的情况下有效管理空间，可以设定两个阈值 rd 和 ri，其中 rd 表示数组元素数量与总空间之比下限，ri 表示数组元素数量与总空间之比上限。当数组元素数量小于 rd 时，将数组大小调整为 ri，这样可以限制额外空间的使用，并确保在一定程度上降低了空间浪费。

    优化额外空间使用：通过合理设置 rd 和 ri 的值，可以控制额外空间的使用，例如可以将额外空间限制在 (1 + ε)n 的范围内，其中ε为任意小于1的正数。

    时间复杂度：动态数组通常支持在末尾进行动态操作的时间复杂度为Θ(1)，如Python中的List的append和pop操作，这是因为它们会根据需要动态调整数组的大小。然而，其他插入和删除操作可能会涉及到移动元素的操作，因此时间复杂度可能为Θ(n)。

总的来说，动态数组的删除操作在末尾进行时具有Θ(1)的时间复杂度，但需要考虑空间的合理利用，通过动态调整数组大小并设置适当的阈值，可以在保持性能的同时有效管理空间。
